# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 清理数据

## 结论要清晰的步骤  
- [x] 数据类型`IncoiceDate`需要转换为日期型，`CustomerID`需要转换为字符串型
- [x] 将`Description`和`CustomerID`的缺失值删除
- [x] 删除Quantity小于0的数据
- [x] 删除UnitPrice小于0的数据
- [x] 对不一致数据将United Kingdom、U.K.替换成UK，将United States 替换成USA

```python
clean_data=original_data.copy()
clean_data
```

```python
clean_data["InvoiceDate"]=pd.to_datetime(clean_data["InvoiceDate"])
clean_data
```

```python
clean_data["CustomerID"]=clean_data["CustomerID"].astype(str).str.slice(0,-2)
```

```python
clean_data.dropna(subset=["Description","CustomerID"],inplace=True)
```

```python
clean_data.isnull().sum()
```

```python
clean_data
```

```python
clean_data["Quantity"]=clean_data["Quantity"][clean_data["Quantity"]>0]
```

```python
clean_data["UnitPrice"]=clean_data["UnitPrice"][clean_data["UnitPrice"]>0]
```

```python
clean_data["Country"]=clean_data["Country"].replace({"United Kingdom":"UK","U.K.":"UK","United States":"USA"})
```

```python
clean_data["Country"].value_counts()
```
